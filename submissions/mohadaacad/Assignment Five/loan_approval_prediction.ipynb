```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression

from sklearn.ensemble import RandomForestClassifier

from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix
)
```

```python
# Step 2: Load Dataset
df = pd.read_csv("C:/Users/mohad/OneDrive/Desktop/Assignment five/clean_loan_dataset.csv")

df.head()
```

```python
print(df.shape)

print(df.info())
```

```python
# Step 3: Prepare Features and Target
X = df.drop(columns=["Approved"])

y = df["Approved"]

print(X.head())

print(y.head())
```

```python
# Step 4: Split Dataset
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Training Features:", X_train.shape)
print("Testing Features :", X_test.shape)

print("Training Target:", y_train.shape)
print("Testing Target :", y_test.shape)
```

```python
log_model = LogisticRegression(
    max_iter=1000,
    random_state=42
)

log_model.fit(X_train, y_train)
```

```python
rf_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

rf_model.fit(X_train, y_train)
```

```python
tree_model = DecisionTreeClassifier(
    random_state=42
)

tree_model.fit(X_train, y_train)
```

```python
def evaluate_model(model, X_test, y_test):

    prediction = model.predict(X_test)

    print("Accuracy :", round(
        accuracy_score(y_test, prediction),3))

    print("Precision:", round(
        precision_score(y_test, prediction),3))

    print("Recall   :", round(
        recall_score(y_test, prediction),3))

    print("F1-Score :", round(
        f1_score(y_test, prediction),3))

    print()

    print("Confusion Matrix")

    print(confusion_matrix(
        y_test,
        prediction
    ))
```

```python
print("Logistic Regression Performance\n")

evaluate_model(
    log_model,
    X_test,
    y_test
)
```

```python
print("Random Forest Performance\n")

evaluate_model(
    rf_model,
    X_test,
    y_test
)
```

```python
print("Decision Tree Performance\n")

evaluate_model(
    tree_model,
    X_test,
    y_test
)
```

```python
sample = X_test.iloc[[0]]

actual = y_test.iloc[0]

log_prediction = log_model.predict(sample)[0]

rf_prediction = rf_model.predict(sample)[0]

tree_prediction = tree_model.predict(sample)[0]

print("Actual Label:")

print(actual)

print()

print("Logistic Regression:")

print(log_prediction)

print()

print("Random Forest:")

print(rf_prediction)

print()

print("Decision Tree:")

print(tree_prediction)
```
