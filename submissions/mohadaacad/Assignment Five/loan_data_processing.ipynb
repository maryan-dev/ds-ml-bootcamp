```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)

df = pd.read_csv("C:/Users/mohad/OneDrive/Desktop/Assignment five/raw_loan_dataset.csv")

print("First 5 Rows")
display(df.head())

print("\nShape:")
print(df.shape)

print("\nDataset Information")
print(df.info())

print("\nMissing Values")
print(df.isnull().sum())
```

```python
print("HasCollateral Value Counts")
print(df["HasCollateral"].value_counts(dropna=False))

print("\nPreviousDefaults Value Counts")
print(df["PreviousDefaults"].value_counts(dropna=False))

print("\nApproved Value Counts")
print(df["Approved"].value_counts(dropna=False))
```

```python
df["Income"] = (
    df["Income"]
    .astype(str)
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
)

df["LoanAmount"] = (
    df["LoanAmount"]
    .astype(str)
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
)

df["Income"] = pd.to_numeric(df["Income"], errors="coerce")
df["LoanAmount"] = pd.to_numeric(df["LoanAmount"], errors="coerce")

print(df[["Income", "LoanAmount"]].head())

print("\nData Types")
print(df[["Income", "LoanAmount"]].dtypes)
```

```python
def clean_yes_no(column):

    return (
        column.astype(str)
        .str.strip()
        .str.lower()
        .replace({
            "yes": "Yes",
            "y": "Yes",
            "true": "Yes",
            "1": "Yes",
            "approved": "Yes",

            "no": "No",
            "n": "No",
            "false": "No",
            "0": "No",
            "rejected": "No",

            "??": np.nan,
            "unknown": np.nan,
            "nan": np.nan
        })
    )
```

```python
columns = [
    "HasCollateral",
    "PreviousDefaults",
    "Approved"
]

for col in columns:
    df[col] = clean_yes_no(df[col])

for col in columns:

    print(f"\n{col}")

    print(df[col].value_counts(dropna=False))
```

```python
# Step 4: Impute Missing Values

# Numeric columns (Median)
df["Income"] = df["Income"].fillna(df["Income"].median())

df["CreditScore"] = df["CreditScore"].fillna(df["CreditScore"].median())

df["EmploymentYears"] = df["EmploymentYears"].fillna(df["EmploymentYears"].median())

df["LoanAmount"] = df["LoanAmount"].fillna(df["LoanAmount"].median())

# Categorical columns (Mode)
df["HasCollateral"] = df["HasCollateral"].fillna(df["HasCollateral"].mode()[0])

df["PreviousDefaults"] = df["PreviousDefaults"].fillna(df["PreviousDefaults"].mode()[0])

df["Approved"] = df["Approved"].fillna(df["Approved"].mode()[0])

print("Missing Values After Imputation")
print(df.isnull().sum())
```

```python
print("Missing Values After Imputation")

print(df.isnull().sum())
```

```python
# Step 5: Remove Duplicate Rows
before_rows = df.shape[0]

df = df.drop_duplicates()

after_rows = df.shape[0]

print("Rows Before:", before_rows)

print("Rows After :", after_rows)

print("Duplicates Removed:", before_rows - after_rows)
```

```python
print("Dataset Shape")

print(df.shape)

display(df.head())
```

```python
# Step 6: Handle Outliers using IQR Capping

outlier_columns = [
    "Income",
    "CreditScore",
    "LoanAmount",
    "EmploymentYears"
]

for col in outlier_columns:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df.loc[:, col] = df[col].clip(lower, upper)

    print(f"\n{col}")
    print("Lower Bound:", lower)
    print("Upper Bound:", upper)
```

```python
# Step 8: Feature Engineering
df.loc[:, "DebtToIncome"] = (
    df["LoanAmount"] /
    (df["Income"] + 1)
)

df.loc[:, "IncomePerYearEmployed"] = (
    df["Income"] /
    (df["EmploymentYears"] + 1)
)
```

```python
# Step 9: Feature Scaling
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

numeric_features = [
    "Income",
    "CreditScore",
    "EmploymentYears",
    "LoanAmount",
    "DebtToIncome",
    "IncomePerYearEmployed"
]

df[numeric_features] = scaler.fit_transform(
    df[numeric_features]
)

print(df[numeric_features].head())
```
